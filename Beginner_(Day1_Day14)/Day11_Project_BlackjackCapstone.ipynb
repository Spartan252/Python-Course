````python
#V1

import  random

import os

def clear_terminal():
    os.system('cls' if os.name == 'nt' else 'clear')

    

def get_cards(cards, times):
    for x in range(times):
        cards.append(random.randint(1,11))
    return cards

def final_score(user, dealer):
    print(f"  Your final hand: {user}, final score: {sum(user)}")
    print(f"  Computer's final hand: {dealer}, final score: {sum(dealer)}")

def blackjack():
    print("""
        .------.            _     _            _    _            _    
        |A_  _ |.          | |   | |          | |  (_)          | |   
        |( \/ ).-----.     | |__ | | __ _  ___| | ___  __ _  ___| | __
        | \  /|K /\  |     | '_ \| |/ _' |/ __| |/ / |/ _' |/ __| |/ /
        |  \/ | /  \ |     | |_) | | (_| | (__|   <| | (_| | (__|   < 
        '-----| \  / |     |_.__/|_|\__,_|\___|_|\_\ |\__,_|\___|_|\_\\
              |  \/ K|                            _/ |                
              '------'                           |__/ 
        """)
    
    user_cards = []
    dealer_cards = []

    get_cards(user_cards, 2)
    print(f"\tYour cards: {user_cards}, current score: {sum(user_cards)}")

    if sum(user_cards) == 21:
            print("You win 游땎")
            final_score(user_cards, dealer_cards)
            check()
    elif (sum(user_cards) > 21):
        print("You went over. You lose 游땴")
        final_score(user_cards, dealer_cards)
        check()


    get_cards(dealer_cards, 2)
    print(f"\tComputer's first card: {dealer_cards[0]}")

    cards = input("Type 'y' to get another card, type 'n' to pass:").lower()

    while cards == "y":
        get_cards(user_cards, 1)
        print(f"\tYour cards: {user_cards}, current score: {sum(user_cards)}")

        if sum(user_cards) == 21:
            print("You win 游땎")
            final_score(user_cards, dealer_cards)
            check()
        elif (sum(user_cards) > 21):
            print("You went over. You lose 游땴")
            final_score(user_cards, dealer_cards)
            check()
        cards = input("Type 'y' to get another card, type 'n' to pass:").lower()

        
    while sum(dealer_cards) < 17 :
        get_cards(dealer_cards, 1, "dealer")


    if (sum(dealer_cards) > 21):
        print("Opponent went over. You win 游때")
        final_score(user_cards, dealer_cards)
        check()
        

    
    if (21 - sum(user_cards)) < (21 - sum(dealer_cards)):
        print("You win 游땎")
    else:
        print("You lose 游땴")
    final_score(user_cards, dealer_cards)
    check()


def check():
    play = input("Do you want to play a game of Blackjack? Type 'y' or 'n': ").lower()
    if play == "y":
        clear_terminal()
        blackjack()



#   Start
user_cards = []
dealer_cards = []
check()
````
