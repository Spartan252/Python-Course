## Concepts

### Conditional operators

**If, elif, else**
```bash
if age >= 12:
    candy == "Lollypop"
elif age < 12:
    candy == "gummybear"
else:
    candy == "chocolate bar"
```


### Comparison Operators

| Operator     | Meaning                                |
|--------------|----------------------------------------|
| >            | Greater than                           |
| <            | Less than                              |
| >=           | Greater than or equal to               |
| <=           | Less than or equal to                  |
| ==           | Equal to                               |
| !=           | Not equal to                           |



### Logical operators
**and, or, not**
```bash
a = 10

print(a > 9 and < 11)       #True
print(a > 9 and == 11)      #False

print(a > 9 or < 11)        #True
print(a > 9 or == 11)       #True
print(a == 9 or == 11)      #False

print(not a == 9)           #True
print(not a == 10)          #False
print(not True)             #False
print(not False)            #True

```

````python
print("Check if the number is odd or even")
number = int(input("What is the number you want to check"))

if number % 2 == 0:
    print(f"The number {number} is even")
else:
    print(f"The number {number} is odd")
````

````python
print("Welcome to the rollercoaster!")

height = float(input("What is your height in cm?\n"))
pay = 0

if height >= 120:
    print("You can ride the rollercoaster")
    age = int(input("What is your age?\n"))

    if age < 12:
        print("Child tickets are $5")
        pay += 5
    elif age < 18 and age >= 12:
        print("Youth tickets are $7")
        pay += 7
    elif age >= 45 and age <= 55:
        print("Everything is going to be okay. Have a free ride on us!")
    else:
        print("Adults tickets are $12")
        pay += 12
    
    photo = input("Do you want a photo of you riding the roallercoaster? (yes/no)").upper()
    if photo == "YES" or "Y":
        pay += 3
        print(f"Please pay ${pay} to ride the roallercoaster")
    else:
        print(f"Please pay ${pay} to ride the roallercoaster")

else:
    print("You can't ride the roallercoaster, you need to be atleast 120cm tall")
````

````python
print("Welcome to the Python Pizza Deliveries!")

bill = 0
size = input("What size pizza d you want? S, M or L\n").upper()
pepperoni = input("Do you want pepperoni in your pizza? Y or N\n").upper()
extra_cheese = input("Do you want extra cheese in your pizza? Y or N\n").upper()

if size == "S":
    bill += 15
elif size == "M":
    bill += 20
elif size == "L":
    bill += 25

if pepperoni == "Y":
    if size == "S":
        bill += 2
    else:
        bill += 3

if extra_cheese == "Y":
    bill += 1

print(f"Your final bill is: ${bill}")
````
