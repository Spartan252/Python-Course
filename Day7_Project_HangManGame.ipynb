````python
import random

lives_display = {
    6: """
  +---+
  |   |
      |
      |
      |
      |
=========
****************************6/6 LIVES LEFT****************************
""",
    5: """
  +---+
  |   |
  O   |
      |
      |
      |
=========
****************************5/6 LIVES LEFT****************************
""",
    4: """
  +---+
  |   |
  O   |
  |   |
      |
      |
=========
****************************4/6 LIVES LEFT****************************
""",
    3: """
  +---+
  |   |
  O   |
 /|   |
      |
      |
=========
****************************3/6 LIVES LEFT****************************
""",
    2: """
  +---+
  |   |
  O   |
 /|\  |
      |
      |
=========
****************************2/6 LIVES LEFT****************************
""",
    1: """
  +---+
  |   |
  O   |
 /|\  |
 /    |
      |
=========
****************************1/6 LIVES LEFT****************************
""",
    0: """
  +---+
  |   |
  O   |
 /|\  |
 / \  |
      |
=========
****************************0/6 LIVES LEFT****************************
"""
}

print(''' _                                             
| |                                            
| |__   __ _ _ __   __ _ _ __ ___   __ _ _ __  
| '_ \ / _' | '_ \ / _' | '_ ' _ \ / _' | '_ \ 
| | | | (_| | | | | (_| | | | | | | (_| | | | |
|_| |_|\__,_|_| |_|\__, |_| |_| |_|\__,_|_| |_|
                    __/ |                      
                   |___/''')


list_of_words = [
    "Run", 
    "Sabotage", 
    "Capoeira", 
    "Elephant", 
    "Mystery", 
    "Rainbow", 
    "Bicycle", 
    "Mountain", 
    "Kitchen", 
    "Freedom", 
    "Penguin", 
    "Adventure", 
    "Chocolate", 
    "Thunder", 
    "Whisper", 
    "Dragon", 
    "Butterfly", 
    "Ocean", 
    "Galaxy", 
    "Phoenix"
]



win = False
game_over = False
lives = 6
guessed_letters = []


random_word = random.choice(list_of_words).upper()


display = []
for letter in random_word:
    display.append("_")


while not game_over:
    print(f"Word to guess: {' '.join(display)}\n")

    user_letter = input("Guess a letter: ").upper()

    if user_letter in guessed_letters:
        print(f"You already guessed {user_letter}")
    else:
        guessed_letters.append(user_letter)


    if user_letter in random_word:
        for position in range(len(random_word)):
            if random_word[position] == user_letter:
                display[position] = user_letter
        print(f"Good guess! {user_letter} is in the word.")
    else:
        lives -= 1
        print(f"You guessed {user_letter}, that's not in the word. You lose a life.")
    print(lives_display[lives])

    if "_" not in display:
        win = True
        game_over = True
    elif lives == 0:
        game_over = True


if win == True:
    print(f"{random_word}")
    print("You win.")
    print(r"""Have your banana
      _
     //\
     V  \
      \  \_
       \,'.`-.
        |\ `. `.       
        ( \  `. `-.                        _,.-:\
         \ \   `.  `-._             __..--' ,-';/
          \ `.   `-.   `-..___..---'   _.--' ,'/
           `. `-_   `._   `-`         /     ,'/'
             \ `-._   `-._        _.-'     ,'/ 
              `-._ `-.___ \     /` _.-'   ,,'
                   `-.__  `----'   _.',-'
                       `--..____..--'
    """)
else:
    print(f"***********************IT WAS {random_word}! YOU LOSE**********************")
    print("The banana rotted away")
````
