## Concepts

### Dictionaries

Keys with associated values

```bash
{key : Value}

programmming_dictionary = {
    "Bug" : "An error in a program that prevents the program from running as expected.",
    "Function" : "A piece of code that you can easily call over and over again.",
    "Loop" : "The action of doing something over and over again."
}

```

**Adding an element:**
```bash
programmming_dictionary = {
    "Bug" : "An error in a program that prevents the program from running as expected.",
    "Function" : "A piece of code that you can easily call over and over again.",
}

programmming_dictionary["Loop"] = "The action of doing something over and over again."

```

**Editing an element:**
```bash
programmming_dictionary = {
    "Bug" : "An error in a program that prevents the program from running as expected.",
    "Function" : "A piece of code that you can easily call over and over again.",
    "Loop" : "The action of doing something over and over again."
}

programmming_dictionary["Bug"] = "A moth in your computer"

```

**Looping in a dictionary:**
```bash
programmming_dictionary = {
    "Bug" : "An error in a program that prevents the program from running as expected.",
    "Function" : "A piece of code that you can easily call over and over again.",
    "Loop" : "The action of doing something over and over again."
}

for key in programmming_dictionary:
    print(key)                                  #Key
    print(programmming_dictionary[key])         #Value

```


### Nesting
```bash
{
    key : [list],
    key2 : {Dict}
}
```

**Nested List in Dictionary:**
```bash
travel_log = {
    "France" : ["Paris", "Lille", "Dijon"],
    "Germany" : ["Stuttgart", "Berlin"]
}

print(travel_log["France"][1])      #print Lille

```

**Nested Dictionary in Dictionary:**
```bash
travel_log = {
    "France" : {
        "num_times_visited" : 8,
        "cities_visited" : ["Paris", "Lille", "Dijon"]
    },
    "Germany" : {
        "num_times_visited" : 3,
        "cities_visited" : ["Stuttgart", "Berlin"]
    }
}

print(travel_log["Germany"]["cities_visited"][0])        #print Stuttgart

```

````python
# The keys in student_scores are the names of the students and the values are their exam scores. 

# Write a program that converts their scores to grades.

# By the end of your program, you should have a new dictionary called student_grades 
# that should contain student names as keys and their assessed grades for values. 

# The final version of the student_grades dictionary will be checked. 


# This is the scoring criteria: 
# - Scores 91 - 100: Grade = "Outstanding" 
# - Scores 81 - 90: Grade = "Exceeds Expectations" 
# - Scores 71 - 80: Grade = "Acceptable" 
# - Scores 70 or lower: Grade = "Fail" 

student_scores = {
    'Harry': 88,
    'Ron': 78,
    'Hermione': 95,
    'Draco': 75,
    'Neville': 60
}

student_grades = {}

for key in student_scores:
    if student_scores[key] >= 91:
        student_grades[key] = "Outstanding"
    elif 90 >= student_scores[key] >= 81:
        student_grades[key] = "Exceeds Expectations"
    elif 80 >= student_scores[key] >= 71:
        student_grades[key] = "Acceptable"
    elif 70 >= student_scores[key]:
        student_grades[key] = "Fail"

print(student_grades)
````

# Day_Project
Create a game based on the user choices, using logical operations for the paths

````python
import os

def clear_terminal():
    os.system('cls' if os.name == 'nt' else 'clear')

def start():
    print("""                         
                           ___________
                                     /
                            )_______(
                            |\"\"\"\"\"\"\"|_.-._,.---------.,_.-._
                            |       | | |               | | '-.
                            |       |_| |_             _| |_..-'
                            |_______| '-'''----------'' '-'
                            )\"\"\"\"\"\"\"(
                           /_________\\
                         .-------------.
                        /_______________\\
    """)
    print("Welcome to the secret auction program.")


bidders = {}
more_bidders = "yes"


while more_bidders == "yes":
    start()
    name = input("What is your name?\n")
    bid = int(input("What is your bid?: $"))
    bidders[name] = bid

    more_bidders = input("Are there any other bidders? Type 'yes or 'no'.\n").lower()
    clear_terminal()
    
winner = max(bidders, key=bidders.get)
value = bidders[winner]

print(f"The winner is {winner} with a bid of ${value}.")
````
