### Importing

 - Basic
```bash
# keyword    Module name
  import     turtle

#Obj    #Class
timmy = turtle.Turtle() 
```

 - from...import...
```bash
# keyword   Module name  keyword   Thing in module
  from      turtle       import    turtle

#Obj    #Class
timmy = Turtle() 
```

 - Aliasing Modules
```bash
# keyword   Module name  keyword   Thing in module
  from      turtle       as        t

#Obj    #Class
timmy = t.Turtle() 
```

### Turtle Challenge 4:
#### Draw a random walk

````python
import turtle
import random

#Obj    #Class
timmy = turtle.Turtle() 
timmy.shape("turtle")           #Change the shape displayed to a turtle
timmy.color("chartreuse")       #Change the color of timmy to green (chartreuse)
timmy.width(15)                 #Change the width of timmy to 15
timmy.speed("fastest")          #Change the speed of timmy to the fastest option

times_to_execute = 200
list_degrees = [0, 90, 180, 270]
list_colors = ["red", "blue", "green", "Gold", "MediumSlateBlue", "Indigo", "MediumVioletRed", "DeepSkyBlue", "PaleGreen", "SpringGreen"]

for times in range(times_to_execute):
    timmy.pencolor(random.choice(list_colors))
    timmy.forward(30)                                   #Move Timmy forward the specified lenght
    timmy.setheading(random.choice(list_degrees))       #Turn right the random amount



my_screen = turtle.Screen()
my_screen.exitonclick()         #Only exits once there is a click on screen
````

### Turtle Excercise:
#### Generate random RGB colors

````python
import turtle
import random

#Obj    #Class
timmy = turtle.Turtle()

timmy.shape("turtle")           #Change the shape displayed to a turtle
timmy.color("chartreuse")       #Change the color of timmy to green (chartreuse)
timmy.width(15)                 #Change the width of timmy to 15
timmy.speed("fastest")          #Change the speed of timmy to the fastest option
turtle.colormode(255)           #Tap into the turtle module (not the object timmy) to accept the RGB values

times_to_execute = 200
list_degrees = [0, 90, 180, 270]
def random_color():
    r = random.randint(0, 255)
    g = random.randint(0, 255)
    b = random.randint(0, 255)
    return (r, g, b)

for times in range(times_to_execute):
    timmy.pencolor(random_color())
    timmy.forward(30)                                   #Move Timmy forward the specified lenght
    timmy.setheading(random.choice(list_degrees))       #Turn right the random amount



my_screen = turtle.Screen()
my_screen.exitonclick()         #Only exits once there is a click on screen
````

### Turtle Challenge 5:
#### Draw a Spirograph

````python
import turtle
import random

circles_needed = int(input("How many circles do you need? (3, 7, 11, 30, etc.)"))
degrees = 360/circles_needed


#Obj    #Class
timmy = turtle.Turtle()

timmy.shape("turtle")           #Change the shape displayed to a turtle
timmy.color("chartreuse")       #Change the color of timmy to green (chartreuse)
timmy.speed("fastest")          #Change the speed of timmy to the fastest option
turtle.colormode(255)           #Tap into the turtle module (not the object timmy) to accept the RGB values


def random_color():
    r = random.randint(0, 255)
    g = random.randint(0, 255)
    b = random.randint(0, 255)
    return (r, g, b)

def draw_spirograph(circles):
    for _ in range(circles):
        timmy.pencolor(random_color())
        timmy.circle(100)           #Draw a circle with a radius of 100
        timmy.right(degrees)        #Turn right the given degrees


draw_spirograph(circles_needed)

my_screen = turtle.Screen()
my_screen.exitonclick()         #Only exits once there is a click on screen
````

````python
#Getting the colors

import colorgram

number_of_colors = 20
tuple_colors = []

colors = colorgram.extract("Day18_PaintingProject.png", number_of_colors)
print(colors)

for color in colors:
    tuple_colors.append((color.rgb.r, color.rgb.g, color.rgb.b))

print()
print(tuple_colors)


#Drawing

import turtle
import random

#Obj    #Class
timmy = turtle.Turtle()

width_of_points = 20

timmy.shape("turtle")           #Change the shape displayed to a turtle
timmy.color("chartreuse")       #Change the color of timmy to green (chartreuse)
timmy.speed("fastest")          #Change the speed of timmy to the fastest option
timmy.width(width_of_points)    #Change the width of timmy to width_of_points
timmy.hideturtle()              #Hide Timmy
turtle.colormode(255)           #Tap into the turtle module (not the object timmy) to accept the RGB values

def move_up(wide):
    timmy.penup()
    timmy.right(180)
    timmy.forward((wide*2)*width_of_points)
    timmy.right(90)
    timmy.forward(width_of_points*2)
    timmy.right(90)

def random_color():
    return random.choice(tuple_colors)


def paint_points(number_of_points):
    wide = int(number_of_points ** 0.5)
    if wide ** 2 < number_of_points:
        wide += 1

    for row in range(wide):
        for col in range(wide):
            if (row * wide + col) >= number_of_points:
                return
            timmy.penup()
            timmy.forward(width_of_points * 2)
            timmy.pendown()
            timmy.dot(wide, random_color())
        move_up(wide)





paint_points(100)

my_screen = turtle.Screen()
my_screen.exitonclick()         #Only exits once there is a click on screen
    
````
